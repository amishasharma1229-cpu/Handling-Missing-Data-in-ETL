Q1. What are the most common reasons for missing data in ETL pipelines?
ANS:

Missing data in ETL (Extract, Transform, Load) pipelines is usually caused by issues in the source systems, data movement process, transformations, or loading stage. The most common reasons include:

1. Source Data Problems
Null or incomplete records in the source database or application.
Data entry errors where users leave fields blank.
System outages preventing data from being captured.
Changes in source schemas (new columns, renamed fields, dropped tables).

2. Extraction Failures
Network interruptions during data extraction.
API rate limits or timeouts causing partial data retrieval.
Permission or authentication issues preventing access to some records.
Incremental extraction errors where change-tracking mechanisms miss updates.


3. Transformation Issues
Data validation rules filtering out records that fail quality checks.
Incorrect joins that exclude unmatched records.
Data type conversion errors (e.g., invalid dates, numeric overflow).
Business logic errors that unintentionally drop data.
Deduplication processes removing valid records mistakenly.

4. Loading Problems
Database constraints (primary keys, foreign keys, NOT NULL constraints) rejecting records.
Batch load failures where only part of the dataset is loaded.
Storage limitations causing incomplete loads.
Transaction rollbacks after errors.

5. Timing and Synchronization Issues
Late-arriving data that hasn't reached the source when extraction occurs.
Out-of-sync systems causing temporary gaps.
Scheduling conflicts between upstream and downstream jobs.

6. Data Quality and Governance Issues
Corrupted files or malformed records.
Inconsistent formats across different sources.
Lack of monitoring and alerting, allowing missing data to go unnoticed.
Poor metadata management, making schema changes difficult to track.


Example

Suppose an ETL pipeline loads customer orders nightly:

The source application adds a new field and changes an existing column name.
The extraction job still expects the old schema.
Records fail validation during transformation and are dropped.
The warehouse ends up with missing order data.

Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?

ans:

Blindly deleting rows with missing values is considered a bad practice in ETL because it can reduce data quality and lead to inaccurate analysis.

Reasons:
Loss of Important Data

Deleting rows may remove valuable information that could be useful for analysis or decision-making.

Reduced Dataset Size

If many rows contain missing values, deleting them can significantly reduce the amount of available data, affecting model performance.
Biased Results

Missing values may follow a pattern (e.g., certain customer groups have more missing data). Removing these rows can introduce bias and produce misleading results.

5.Lower Accuracy

Machine learning models and reports built on a smaller dataset may become less accurate and less representative of the real-world data.

Not All Missing Values Are Important

Sometimes only one non-critical column is missing. Deleting the entire row wastes other valid information in that record.

Q3. Explain the difference between:

Listwise deletion

Column deletion

Listwise Deletion


Removes entire rows (records) that contain one or more missing values.
Used when only a small number of rows have missing data.
May reduce the number of records available for analysis.
Can introduce bias if many rows are deleted.


Column Deletion


Removes entire columns (features) that contain many missing values.

Used when a column has too many missing values and provides little useful information.

May reduce the number of features available for analysis.

Can lead to loss of an important variable if the deleted column is useful.

Q4. Why is median imputation preferred over mean imputation for skewed data such as income?

**Median imputation** is preferred over **mean imputation** for **skewed data** (such as income) because the median is **less affected by extreme values (outliers)**.

### Reasons:

1. **Resistant to Outliers**

   * In skewed datasets, a few very high or very low values can significantly change the mean.
   * The median remains stable because it is the middle value of the sorted data.

2. **Better Represents the Typical Value**

   * The median reflects the central tendency more accurately for skewed distributions, making it a better replacement for missing values.

3. **Preserves Data Distribution**

   * Median imputation maintains the original distribution of the data better than mean imputation, which may shift the average toward extreme values.

### Example:

Income values: **₹20,000, ₹25,000, ₹30,000, ₹35,000, ₹5,00,000**

* **Mean = ₹1,22,000** (heavily affected by the outlier ₹5,00,000)
* **Median = ₹30,000** (represents a typical income)

If an income value is missing, replacing it with **₹30,000 (median)** is more realistic than **₹1,22,000 (mean)**.

**Conclusion:**
For skewed data such as income, **median imputation** is preferred because it is robust to outliers, better represents the typical value, and preserves the data distribution more effectively than mean imputation.





Q5. What is forward fill and in what type of dataset is it most useful?

**Forward Fill** (also called **Forward Propagation** or **ffill**) is a missing value handling technique in which each missing value is replaced with the **last valid (non-missing) value** that appears before it.

### When is Forward Fill Most Useful?

Forward fill is most useful for **time-series** or **sequential datasets**, where observations are recorded in chronological order and the previous value is expected to remain valid until a new value is available.

### Example:

| Date  | Temperature |
| ----- | ----------- |
| 1 Jan | 25°C        |
| 2 Jan | *Missing*   |
| 3 Jan | *Missing*   |
| 4 Jan | 28°C        |

After applying **Forward Fill**:

| Date  | Temperature |
| ----- | ----------- |
| 1 Jan | 25°C        |
| 2 Jan | 25°C        |
| 3 Jan | 25°C        |
| 4 Jan | 28°C        |

### Advantages:

* Simple and fast to implement.
* Preserves the continuity of sequential data.
* Prevents data loss caused by missing values.

### Limitation:

* It assumes the previous value remains unchanged until the next recorded value, which may not always be true.

**Conclusion:**
Forward fill is a technique that replaces missing values with the previous valid value. It is most effective for **time-series and sequential datasets**, such as stock prices, weather data, sensor readings, and sales records.


Q6. Why should flagging missing values be done before imputation in an ETL workflow?

**Flagging missing values** means creating a new indicator column (e.g., `Age_Missing`) that records whether a value was originally missing before it is filled using an imputation technique.

### Why should it be done before imputation?

1. **Preserves Information**

   * Once missing values are imputed, it becomes impossible to identify which values were originally missing.
   * A flag keeps this information for future analysis.

2. **Improves Machine Learning Models**

   * Missingness itself can carry useful information.
   * Models can use the flag column as an additional feature to improve prediction accuracy.

3. **Supports Data Quality Analysis**

   * Analysts can measure how much data was missing and identify patterns of missing values.

4. **Enables Auditing and Traceability**

   * ETL processes become more transparent because users can distinguish between original and imputed values.

### Example:

**Original Data**

| CustomerID | Age       |
| ---------- | --------- |
| 101        | 25        |
| 102        | *Missing* |
| 103        | 30        |

**After Flagging and Median Imputation**

| CustomerID | Age | Age_Missing |
| ---------- | --- | ----------- |
| 101        | 25  | 0           |
| 102        | 27  | 1           |
| 103        | 30  | 0           |

Here, `Age_Missing = 1` indicates that the original value was missing and later imputed.

**Conclusion:**
Flagging missing values before imputation preserves the original missing-data information, improves data quality analysis, supports auditing, and can enhance the performance of machine learning models.


### Q7. Consider a scenario where income is missing for many customers.

If the **Income** column has missing values for many customers, simply deleting those records can lead to a significant loss of data and may introduce bias. Instead, appropriate missing-value handling techniques should be applied.

### Recommended Approach

1. **Analyze the Missing Data**

   * Determine how many income values are missing and identify any patterns in the missing data.

2. **Choose a Suitable Imputation Method**

   * If the income data is **skewed** (which is common), use **median imputation** because the median is less affected by extreme values.
   * If the data is approximately **normally distributed**, **mean imputation** may be used.

3. **Flag Missing Values**

   * Create a new column such as **Income_Missing** to indicate whether the original income value was missing before imputation.

4. **Validate the Results**

   * Check that the imputed values are reasonable and do not significantly alter the overall data distribution.

### Example

**Original Data**

| CustomerID | Income    |
| ---------- | --------- |
| 101        | ₹30,000   |
| 102        | *Missing* |
| 103        | ₹45,000   |
| 104        | *Missing* |

**After Median Imputation and Flagging**

| CustomerID | Income  | Income_Missing |
| ---------- | ------- | -------------- |
| 101        | ₹30,000 | 0              |
| 102        | ₹37,500 | 1              |
| 103        | ₹45,000 | 0              |
| 104        | ₹37,500 | 1              |

### Conclusion

When income is missing for many customers, it is better to **impute the missing values (preferably using the median for skewed data), create a missing-value flag, and avoid deleting records**. This preserves valuable data, reduces bias, and improves the quality of analysis and machine learning models.
